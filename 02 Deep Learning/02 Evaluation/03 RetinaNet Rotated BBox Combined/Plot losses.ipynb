# Losses

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


def plot_losses(loss_list, legends, title='', ylabel='Loss'):
    plt.figure(figsize=(4, 3), dpi=200)
    
    for loss in loss_list:
        plt.plot(list(range(1, 1 + len(loss))), loss)
    plt.title(title)
    plt.legend(legends)
    plt.ylabel(ylabel)
    plt.xlabel('Epoch')
    return
```

```python
data = np.genfromtxt('Rotated_Comb_OP_rectified_train_trns_set_13-05_00_losses.csv', delimiter=',')
data2 = []# np.genfromtxt('Rotated_Comb_OP_train_set_14-04_rndTrans_BS4_25-50_losses.csv', delimiter=',')

if len(data2) > 0 :
    val_loss = np.append(data[1::, 0],data2[1::, 0])
    val_regression_loss  = np.append(data[1::, 1],data2[1::, 1])
    val_classification_loss  = np.append(data[1::, 2],data2[1::, 2])
    loss  = np.append(data[1::, 3],data2[1::, 3])
    regression_loss  = np.append(data[1::, 4],data2[1::, 4])
    classification_loss  = np.append(data[1::, 5],data2[1::, 5])
    mAP  = np.append(data[1::, 6],data2[1::, 6])
    lr  = np.append(data[1::, 7],data2[1::, 7])
else:
    val_loss = data[1::, 0]
    val_regression_loss  = data[1::, 1]
    val_classification_loss  = data[1::, 2]
    loss  = data[1::, 3]
    regression_loss  = data[1::, 4]
    classification_loss  = data[1::, 5]
    mAP  = data[1::, 6]
    lr  = data[1::, 7]
```

### Combined Loss

```python
legends = ['train', 'val']
title = 'Combined Loss'
plot_losses([loss, val_loss], legends, title)
```

### Regression Loss

```python
legends = ['train','val']
title = 'Regression BB Loss'
plot_losses([regression_loss, val_regression_loss], legends, title)
```

### Classification Loss

```python
legends = ['train', 'val']
title = 'Classification Loss'
plot_losses([classification_loss, val_classification_loss], legends, title)
```
